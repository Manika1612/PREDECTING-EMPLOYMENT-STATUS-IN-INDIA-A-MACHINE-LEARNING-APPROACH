```python
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDRegressor, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder

```

```python
# Split X and Y
X = df_filtered.drop('Employment_status', axis=1)
Y = df_filtered['Employment_status']
```

```python
#Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=100)
```

```python
scaler = StandardScaler()
X_train[['b4q6_perv1', 'b4q10_perv1']] = scaler.fit_transform(X_train[['b4q6_perv1', 'b4q10_perv1']])
X_test[['b4q6_perv1', 'b4q10_perv1']] = scaler.transform(X_test[['b4q6_perv1', 'b4q10_perv1']])
```

```python
# Train logistic regression
label_encoder = LabelEncoder()
# Fit and transform the 'qtr_perv1' column in both training and testing sets
X_train['qtr_perv1'] = label_encoder.fit_transform(X_train['qtr_perv1'])
X_test['qtr_perv1'] = label_encoder.transform(X_test['qtr_perv1'])


log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, Y_train)

#Predict and evaluate
y_pred_log = log_reg.predict(X_test)
print("Logistic Regression:\n", classification_report(Y_test, y_pred_log))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# Get confusion matrix
cm = confusion_matrix(Y_test, y_pred_log)

# Plot it
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=log_reg.classes_)
disp.plot(cmap='Reds')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()
```

```python
num_features = ['b4q6_perv1', 'b4q10_perv1']
cat_features = ['b4q5_perv1', 'b4q7_perv1', 'b4q8_perv1', 'b4q9_perv1', 'b1q3_perv1', 'qtr_perv1']
```

```python
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)])
```

```python
from sklearn.ensemble import RandomForestClassifier
# Random Forest
pipeline_rf = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', RandomForestClassifier(n_estimators=100, max_depth=10, n_jobs=-1, random_state=42))
])

rf_model = RandomForestClassifier(max_depth=10, random_state=42)
pipeline_rf.fit(X_train, Y_train)
y_pred_rf = pipeline_rf.predict(X_test)
print("Random Forest:\n", classification_report(Y_test, y_pred_rf))
```

```python

# Decision Tree
from sklearn.tree import DecisionTreeClassifier
pipeline_dt = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', DecisionTreeClassifier(max_depth=10, random_state=42))
])

pipeline_dt.fit(X_train, Y_train)
y_pred_dt = pipeline_dt.predict(X_test)
print("Decision Tree Classification Report:\n")
print(classification_report(Y_test, y_pred_dt))
```

```python
num_features = ['b4q6_perv1', 'b4q10_perv1']
cat_features = ['b4q5_perv1', 'b4q7_perv1', 'b4q8_perv1', 'b4q9_perv1', 'b1q3_perv1', 'qtr_perv1']

# Column Transformer
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
])

# Pipeline
clf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(
        max_depth = 4,
        criterion='gini',
        random_state=42
    ))
])

# Train
clf_pipeline.fit(X_train, Y_train)


```

```python
print(f"Tree Depth: {clf_pipeline.named_steps['classifier'].get_depth()}")
print(f"Number of Leaves: {clf_pipeline.named_steps['classifier'].get_n_leaves()}")
```
